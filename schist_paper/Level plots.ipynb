```python
import scanpy as sc
import schist as scs
%pylab
%matplotlib inline
```

```python
sc.set_figure_params(dpi_save=300)
```

```python
adata = scs.io.read("Tian/Tian_pc50_n15_dcTrue")

name = 'sc-mixology'
ann = 'cell_line'
lk = 1
lq = 1
k = 21

level = lk
#adata.uns[f'nsbm_level_{level}_colors'] = [mpl.colors.to_hex(x) for x in cm.Set2(range(len(adata.obs[f'nsbm_level_{level}'].cat.categories)))]
sc.pl.umap(adata, color=f'nsbm_level_{level}', 
           title='', 
           frameon=False, legend_loc='none',
          save=f"{name}_K{lk}.png")

level = lq
#adata.uns[f'nsbm_level_{level}_colors'] = [mpl.colors.to_hex(x) for x in cm.Set3(range(len(adata.obs[f'nsbm_level_{level}'].cat.categories)))]
sc.pl.umap(adata, color=f'nsbm_level_{level}', 
           title='', 
           frameon=False, legend_loc='none',
          save=f"{name}_Q{lq}.png")

sc.pl.umap(adata, color=f'{ann}', 
           title='', 
           frameon=False, legend_loc='none',
          save=f"{name}_Annotation.png")

g = adata.uns['schist']['state'].g
B = gt.BlockState(g, b=adata.obs[ann].cat.codes.values)
M = gt.modularity(g, B.get_blocks())

levels = adata.obs.filter(like='nsbm_level').columns
eff_level = np.sum([adata.obsm[f'CM_{l}'].shape[1]  > 1 for l in levels]) + 1

plot(adata.uns['schist']['stats']['modularity'])
xticks(range(eff_level + 1))
xlim(0, eff_level)
ylabel('$Q$')
xlabel("nSBM level")
grid()
ylim(0, 1)
hlines(M, 0, eff_level, color='C1', linestyle='--')
savefig(f"{name}_modularity.png")
```
